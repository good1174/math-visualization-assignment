```python
import numpy as np
import matplotlib.pyplot as plt
import math

X = np.linspace(-10,10,200)
y = X
y2 = X*X
ysin = np.sin(X)
y4 = np.exp(-0.1 * X) * np.cos(X)
fig, axes = plt.subplots(2,2,figsize=(10,6))
axes[0,0].plot(X,y,label='y=x',marker='o')
axes[0,0].set_title('y=x graph')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')
axes[0,0].legend()
axes[0,0].grid(True)
axes[0,1].plot(X,y2,label='y=x^2',marker='s')
axes[0,1].set_title('y=x^2 graph')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')
axes[0,1].legend()
axes[0,1].grid(True)
axes[1,0].plot(X,ysin,label='y=sin(x)',marker='d')
axes[1,0].set_title('y=sin(x) graph')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')
axes[1,0].legend()
axes[1,0].grid(True)
axes[1,1].plot(X, y4, label='y=e^(-0.1x) * cos(x)',marker='*')
axes[1,1].set_title('y=e^(-0.1x) * cos(x) graph')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')
axes[1,1].legend()
axes[1,1].grid(True)
plt.show()
plt.savefig('function_plot.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

X = np.linspace(-10,10,200)
y = np.cos(X)*np.sin(X) + X
fig, axes = plt.subplots(1,1,figsize=(10,6))
axes.plot(X,y,label='y=x')
axes.set_title('y=cos(x)*sin(x)+x')
axes.set_xlabel('x')
axes.set_ylabel('y=cos(x)*sin(x)+x')
axes.legend()
axes.grid(True)
plt.show()
plt.savefig('own_equation.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

students = ["S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9", "S10"]
midterm = [85, 72, 90, 66, 78, 92, 60, 74, 88, 95]
final = [80, 70, 94, 68, 75, 90, 65, 72, 84, 96]
total_scores = [0.4* m + 0.6* f for m, f in zip(midterm, final)]

plt.figure(figsize=(8, 6))
plt.scatter(midterm, final)
plt.xlabel('Midterm score')
plt.ylabel('Final score')
plt.grid(True)
plt.savefig('score_scatter.png')
plt.show()

plt.figure(figsize=(8, 6))
plt.bar(students, total_scores)
plt.xlabel('Student')
plt.ylabel('Total score')
plt.grid(True)
plt.savefig('score_histogram.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(students, total_scores)
plt.title('Student total score')
plt.xlabel('Student')
plt.ylabel('Total Score')
plt.grid(axis='y')
plt.xticks(rotation=45, ha='right')
plt.savefig('score_bar_chart.png')
plt.show()

```

```python
slope, intercept = np.polyfit(midterm, final, 1)
predicted_final = slope * np.array(midterm) + intercept

plt.figure(figsize=(8, 6))
plt.scatter(midterm, final, label='Original')
plt.plot(midterm, predicted_final, label=f'Predicted')

plt.title('Assignment4')
plt.xlabel('Midterm Score')
plt.ylabel('Final Score')
plt.grid(True)
plt.legend()
plt.savefig('score_prediction.png')
plt.show()

def predicted(x):
  slope*np.array(x) + intercept
  return slope*np.array(x) + intercept

print(predicted(50))
print(predicted(75))
print(predicted(100))
```
